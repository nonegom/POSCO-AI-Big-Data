```python
import pandas as pd
```

```python
meta = pd.read_csv('./Data_OTT/movie_meta.csv', engine = 'python')
down = pd.read_csv('./Data_OTT/movie_down.csv', engine = 'python')
inventory = pd.read_csv('./Data_OTT/movie_inventory1.csv', engine = 'python')
inventory.drop(['Unnamed: 0'], axis=1, inplace = True)

display(meta.head())
display(down.head())
display(inventory.head())
```

```python
new_down = pd.merge(down, inventory, on = 'item_id')
new_down.head()
```

```python
import seaborn as sns

#제일 많이 본 studio
display(new_down.groupby('studio').count()[['item_id']].sort_values(by = 'item_id', ascending=False))
sns.countplot(new_down['studio'])
```

```python
#제일 많이 본 장르
display(new_down.groupby('Genre_1').count()[['item_id']].sort_values(by = 'item_id', ascending=False))
sns.countplot(new_down['Genre_1'])
```

```python
#제일 많이 본 장르
display(new_down.groupby('director').count()[['item_id']].sort_values(by = 'item_id', ascending=False))
sns.countplot(new_down['director'])
```
