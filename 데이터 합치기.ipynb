```python
import pandas as pd
```

# 1. 데이터 불러오기

```python
# movie_customer 데이터 불러오기
df_mc = pd.read_csv('movie_customer.csv', encoding = 'euc-kr')
print(df_mc.shape)
df_mc.head()
```

```python
# movie_down 데이터 불러오기
df_md = pd.read_csv('movie_down.csv', encoding = 'euc-kr')
print(df_md.shape)
df_md.head()
```

```python
# movie_inventory 데이터 불러오기
df_mi = pd.read_csv('movie_inventory.csv', encoding = 'euc-kr')
print(df_mi.shape)
df_mi.head()
```

```python
# movie_meta 데이터 불러오기
df_mm = pd.read_csv('movie_meta.csv', encoding = 'euc-kr')
print(df_mm.shape)
df_mm.head()
```

# movie_down과 movie_customer 결합

```python
# movie_dow과 nmovie_customer 결합
# customer_id를 기준
# 왼쪽 데이터 고유값 기준 (left)
df_mc_md_outer = pd.merge(df_md, df_mc, how = 'left', on = ['customer_id'])
df_mc_md_outer
```

```python
# 결측치 존재하면 제거
df_mc_md = df_mc_md_outer.dropna()
df_mc_md
```

# movie_down, movie_customer와 movie_inventory 결합

```python
# movie_down, movie_customer와 movie_inventory 결합
# item_id를 기준
# 왼쪽 데이터 고유값 기준 (left)
df_mc_md_mi_outer = pd.merge(df_mc_md, df_mi, how = 'left', on = ['item_id'])
df_mc_md_mi_outer
```

```python
# 결측치 존재하면 제거
df_mc_md_mi = df_mc_md_mi_outer.dropna()
df_mc_md_mi
```

# movie_down, movie_customer, movie_inventory와 movie_meta 결합

```python
# movie_down, movie_customer, movie_inventory와 movie_meta 결합
# movie_id 기준
# 왼쪽 데이터 고유값 기준 (left)
df_mc_md_mi_mm_outer = pd.merge(df_mc_md_mi, df_mm, how = 'left', on = ['movie_id'])
df_mc_md_mi_mm_outer
```

```python
# 결측치 존재하면 제거
df_mc_md_mi_mm = df_mc_md_mi_mm_outer.dropna()
df_mc_md_mi_mm
```
