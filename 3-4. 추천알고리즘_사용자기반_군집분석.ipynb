```python
studio_name = down_20_inv['studio'].unique()
studio_num = [i for i in range(len(studio_name))]

studio = pd.DataFrame({'studio' : studio_name,
                       'studio_num' : studio_num})
studio
```

```python
director_name = down_20_inv['director'].unique()
director_num = [i for i in range(len(director_name))]

director = pd.DataFrame({'director' : director_name,
                        'director_num' : director_num})

director
```

```python
genre_name = down_20_inv['Genre_1'].unique()
genre_num = [i for i in range(len(genre_name))]

genre = pd.DataFrame({'Genre_1' : genre_name,
                     'genre_num' : genre_num})

genre
```

```python
down_20_inv = pd.merge(down_20_inv, studio, on = 'studio')
down_20_inv = pd.merge(down_20_inv, director, on = 'director')
down_20_inv = pd.merge(down_20_inv, genre, on = 'Genre_1')
down_20_inv.head()
```

```python
from sklearn_extra.cluster import KMedoids
```

```python
n_iter = 15
cluster_range = [i+1 for i in range(n_iter)]
clus_error = []

for v in cluster_range:
    clus = KMedoids(n_clusters=v)
    clus.fit(down_20_inv[['studio_num', 'director_num', 'genre_num']])
    clus_error.append(clus.inertia_)

ds_error = pd.DataFrame({'NumberofCluster':cluster_range,
                       'Error':clus_error})
ds_error
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 10))
plt.plot(ds_error['NumberofCluster'], ds_error['Error'])
plt.title('Sum of squared distance')
```

```python
clus = KMedoids(6)
clus.fit(down_20_inv[['studio_num', 'director_num', 'genre_num']])

cluster_kmeans = [i+1 for i in clus.labels_]
down_20_inv["ClusterKmeans"] = cluster_kmeans
down_20_inv.head()
```
